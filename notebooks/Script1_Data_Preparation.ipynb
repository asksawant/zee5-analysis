```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('/content/drive/MyDrive/ml-workspace/zee5/data/zee_tv_show.csv')
```

```python
df.head(5)
```

```python
# dimension of the dataset
df.shape
```

```python
# removing the rows which show_name as NA
df = df[df['show_name'].notnull()]
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df['channel'].value_counts()
```

```python
df['genres'].value_counts()
# we can melt this
```

```python
# create an Id column
# which uniquely identify each tv show
df.reset_index(inplace=True)
```

```python
# Rename the index column as id
df.rename(columns = {'index':'id'},inplace=True)
```

```python
df.head()
```

```python
# Creating the fact table
Fact_tv_show = df[['id','show_name','channel','release_date','total_episodes']]
Fact_tv_show.head(3)
```

```python
# Dim_cast
Dim_cast = df[['id','cast_names']]
Dim_cast.head()
```

```python
from ast import literal_eval

Dim_cast['cast_names'] = Dim_cast['cast_names'].apply(literal_eval) #convert to list type
Dim_cast = Dim_cast.explode('cast_names')
Dim_cast.reset_index(drop=True,inplace=True)
```

```python
Dim_cast['cast_names'] = Dim_cast['cast_names'].str.strip()
Dim_cast = Dim_cast[Dim_cast['cast_names'].notnull()]
Dim_cast
```

```python
# Dim_genres
Dim_genres = df[['id','genres']]
Dim_genres.head()
```

```python
# use str split on the column
Dim_genres['genres'] = Dim_genres['genres'].str.split(',')
```

```python
Dim_genres.head()
```

```python
Dim_genres = Dim_genres.explode('genres')
Dim_genres.reset_index(drop=True,inplace=True)
Dim_genres['genres'] = Dim_genres['genres'].str.strip()
Dim_genres = Dim_genres[Dim_genres['genres'].notnull()]
Dim_genres
```

```python
Dim_genres['genres'].value_counts()
```

```python
Fact_tv_show.to_csv('/content/drive/MyDrive/ml-workspace/zee5/data/power_bi/Fact_tv_show.csv',index=False)
Dim_cast.to_csv('/content/drive/MyDrive/ml-workspace/zee5/data/power_bi/Dim_cast.csv',index=False)
Dim_genres.to_csv('/content/drive/MyDrive/ml-workspace/zee5/data/power_bi/Dim_genres.csv',index=False)
```
